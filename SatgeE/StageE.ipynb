```python
import re
def overlapping_chunks(text,max_tokens=300,overlapping_factor=5):
  sentences=re.split(r'(?<=[.?!])\s',text)
  n_tokens==0
  chunks=[]
  for sentence in sentences:
    tokens=tokenizer.encode(sentence)
    if n_tokens+len(tokens)<=max_tokens:
      chunks.append(sentence)
      n_tokens+=len(tokens)
      if overlapping_factor>0:
        for i in range(1,overlapping_factor+1):
          if n_tokens+len(tokens)<=max_tokens:
            chunks.append(sentence)
            n_tokens+=len(tokens)
          else:
            chunks=[]
            n_tokens=0
    if n_tokens>max_tokens:
      continue

    chunks.append(sentence)
    n_tokens+=tokens+1
  if chunks:
    chunks.append(".".join(chunks)+".")
  return chunks
```
